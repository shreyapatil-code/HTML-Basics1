# HTML Basics

1.  Explain the basic structure of an HTML page. Why is it important to follow
this structure?
- Basic Structure of HTML page:


```python
'''
Basic Structure of html page:
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <title>Page title</title>
</head>
<body>
  <!-- website content -->
</body>
</html>
'''
```

Explaination of each part:
- Doctype html: Declares the document type as html5.
- html: It is the root element of html page and all other elements are placed inside it.
- head: Contains meta data(information about the page) and it is not visible to users.
- title: present inside the head tag and is shown in browser tab.
- body: contains all visible contents such as headings, paragraphs, images etc.
* Importance of HTML Structure:
1. Browsers understand the page correctly and display content without errors.
2. Keeps code clean, structured, and easy to maintain.
3. Ensures the page works consistently on different browsers.
4. Using proper meta tags helps pages adapt to different screen sizes.
5. Search engines can better understand and rank the page.

3. You’re creating a form that asks the user to enter their name, select their
country from a list, and click a button to submit. Write the HTML code for this.

```python
'''
<form>
  <label for="name">Enter Name:</label>
  <input type="text" id="name" name="name" placeholder="Enter your name"><br><br>

  <label for="country">Select Country:</label>
  <select id="country" name="country">
    <option value="India">India</option>
    <option value="USA">USA</option>
    <option value="Russia">Russia</option>
    <option value="France">France</option>
    <option value="Japan">Japan</option>
  </select><br><br>

  <input type="submit" value="Submit">
</form>
'''
```

5. Write short HTML code to show your favorite fruits in a bulleted list and
your top 3 programming languages in a numbered list.

```python
'''
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>My Lists</title>
</head>

<body>
<h2>My favourite fruits list</h2>
  <ul>
    <li>Mango</li>
    <li>Apple</li>
    <li>Strawberry</li>
    <li>Watermelon</li>
    <li>Guava</li>
  </ul>
<h2>Top 3 programming languages:</h2>
  <ol>
     <li>javascript</li>
     <li>python</li>
     <li>C</li>
  </ol>
</body>
</html>
'''
```

6.  Ravi wants to leave a note in his HTML file so that only other developers

useful?
